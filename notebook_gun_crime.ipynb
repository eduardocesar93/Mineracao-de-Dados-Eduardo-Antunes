<i> Importando bibliotecas </i>

```python
import os
import sys
import time
import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

<i> Dados disponíveis no repositório <a href="https://github.com/jldbc/gunsandcrime">gunsandcrime</a></i>

```python
datafile = "gun_crime_data.csv"
```

<b>Explorando o dataset</b>

```python
df = pd.read_csv(datafile)
df = df.dropna() # Retirando dados com valores NaN
df = df.sort_values(['year']) # Ordenando crescentemente por ano, para melhor visualização das séries temporais
df.head()
```

```python
df.describe()
```

Dados de Assassinato e Estupro

```python
df.plot('year', 'murder')
df.plot('year', 'rape')
```
